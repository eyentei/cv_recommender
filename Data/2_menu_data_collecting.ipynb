```python
import requests
from bs4 import BeautifulSoup

# ищем ссылки, которые относятся к категориям 
html = requests.get('https://www.mcdonalds.com/us/en-us/full-menu.html').text
bs = BeautifulSoup(html)
res = bs.findAll("a", { "class" : "category-link" })
```

```python
links_to_categories = []
for i in res:
    links_to_categories.append('https://www.mcdonalds.com'+i.attrs['href'])  
    # cоздаем список ссылок на категории
```

```python
links_to_categories
```

```python
menu_items = []
for link in links_to_categories:
    html = requests.get(link).text
    bs = BeautifulSoup(html)
    res = bs.findAll("a", { "class" : "categories-item-link" })
    for i in res:
        menu_items.append('https://www.mcdonalds.com'+i.attrs['href'])  
        # cоздаем список ссылок на продукты
```

```python
menu_items
```

```python
mac_menu = []
for item in menu_items:

    html = requests.get(item).text
    bs = BeautifulSoup(html)

    title = bs.findAll("h1")

    res = bs.findAll("span", { "class" : "ingredient-text" })
    
    img  = bs.findAll("img", {'class': 'img-responsive cont-100'})
    
    if len(img)>0 and len(title)>0:
        img_data = requests.get('https://www.mcdonalds.com'+img[0]['srcset']).content
        with open(f'images/{title[0].text}.jpg', 'wb') as f:
            f.write(img_data)
    
        ingredients = set()
    
        for span in res:
            ingredients.add(span.text)

        mac_menu.append({'item':title[0].text, 'ingredients':list(ingredients)})
    
    # со страниц продуктов достаем названия блюд и их составы
```

```python
import pandas as pd
mac = pd.DataFrame(mac_menu)
```

```python
mac.sort_values("item", inplace = True) 
mac
```

```python
dr = mac.drop_duplicates(subset ="item") 
dr
```

```python
dr = dr.reset_index(drop=True)
```

```python
dr.to_csv('mac_menu.csv')
```

```python
dr.ingredients.apply(pd.Series).merge(dr, left_index = True, right_index = True) # разобьем ингредиенты по столбцам
```

```python
new = dr.ingredients.apply(pd.Series) \
    .merge(dr, right_index = True, left_index = True) \
    .drop(["ingredients"], axis = 1) \
    .melt(id_vars = ['item'], value_name = "ingredient") \
    .drop("variable", axis = 1)

# приведем к виду many-to-many таблицы
```

```python
new.to_csv('mac_m.csv',sep = ';',index=None)
```

```python
# теперь поместим полученные данные в базу
import pandas as pd
import psycopg2 as pg
with open('credentials.txt') as creds:
    db,user,host,port = creds.read().splitlines()
connection = pg.connect(dbname=db, user=user, host=host, port=port)
cursor = connection.cursor()
```

```python
menu=pd.read_csv('mac_m.csv',sep = ';')
```

```python
items = list(menu.item.unique())
tags = list(menu.tags.unique())
```

```python
# записываем теги
for tag in tags:
    cursor.execute("insert into tags (tag_name) values (%s)", (tag,))
connection.commit()
```

```python
# записываем продукты
for item in items:
    cursor.execute("insert into items (item_name) values (%s)", (item,))
connection.commit()
```

```python
# соотносим продукты с тегами
for i, pair in menu.iterrows():
    data = (items.index(pair['item'])+1,tags.index(pair['tags'])+1)
    cursor.execute("insert into items_tags (item_id, tag_id) values (%s, %s)", data)

connection.commit()
```
